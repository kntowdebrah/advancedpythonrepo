```python
#changing colors 
import matplotlib.pyplot as plt 
import numpy as np 

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c= colors)

plt.show()
```

```python
#ColorMap
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75]) # size of each can be specified or default when removed. 

plt.scatter(x, y, s=sizes, c = colors, cmap = 'viridis', alpha=0.5) #alpha attribute can be removed in which case the default holds. 

plt.colorbar() #can be removed

plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

#startangle specifies where to start drawing the pie chart instead of the default zero
# explode attribute specifies which piece of the pie will be separated from the whole
# we can add the shadow attribute to it as well
plt.pie(y, labels = mylabels, startangle= 90, explode=myexplode, shadow=True) 
plt.show()
```

```python
#colors
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.legend(title ="Four Fruits:", loc ="upper left")
plt.show()
```
